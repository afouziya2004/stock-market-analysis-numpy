```python
stock_data = np.genfromtxt(
    "nse_stock_data.csv",
    delimiter=",",
    skip_header=1
)

stock_data
```

```python
close_prices = stock_data[:, 3]
```

```python
plt.plot(close_prices)
plt.title("Stock Close Price Trend")
plt.xlabel("Days")
plt.ylabel("Close Price")

plt.show()
```

```python
returns = np.diff(close_prices) / close_prices[:-1]
returns
```

```python

plt.figure()
plt.plot(returns)
plt.title("Daily Returns")
plt.xlabel("Days")
plt.ylabel("Returns")
plt.tight_layout()

plt.savefig("daily_returns.png")
plt.show()
```

```python
np.mean(close_prices)
np.std(close_prices)
np.var(close_prices)
```

```python
volatility = np.std(returns)
volatility
```

```python
close_prices = stock_data[:, 3]
plt.figure()
plt.hist(close_prices, bins=6)
plt.title("Closing Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.tight_layout()

plt.savefig("price_distribution.png")
plt.show()
```
